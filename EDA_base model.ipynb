1. 배경

최근 인류에게 다가온 빅데이터라는 단어는 우주와 천문학에게 낯설지 않습니다.  
찰나의 순간에도 우주는 천문학적인 양의 데이터를 생산해왔고, 오래 전부터 천문학자들은 우주를 관측했으며, 그 방대함에 비례하는 데이터를 수집 및 분석했기 때문입니다. 
슬론 디지털 천체 관측(Sloan Digital Sky Survey: 이하 SDSS)는 세계적 천체 관측 프로젝트로, 우주에 대한 천문학적인 규모의 데이터를 수집하고 있습니다.   
이곳에서 수집한 데이터는 약 6,000개 논문에 사용되었고, 25만 회 이상 인용되었을 정도로 천문학에 큰 기여를 했습니다.  
점점 거대해지는 규모에 따라 데이터 처리에는 머신러닝과 딥러닝 기법이 활용되기 시작했습니다.  
여전히 우주에는 다양한 미지의 이야기가 남아있고, 오늘날 인간은 하늘에서 많은 데이터를 얻어낼 정도로 발전했습니다.  
이 데이터를 분석하여 어쩌면 드러나지 않은 규칙이 여러분의 손끝에서 밝혀질 수 있습니다.   
새로운 알고리즘을 통해 우주의 비밀을 찾아주세요!

3. 컬럼 설명
    * type(Source type) : 천체의 분류 예측을 해야하는 변수
    * psfMag(Point spread function magnitudes): 먼 천체를 한 점으로 가정하여 측정한 빛의 밝기입니다.

    * modelMag(Model magnitudes) : 천체 중심으로부터 특정 거리의 밝기입니다.
    * fiberID : 관측에 사용된 광섬유의 구분자

# Baseline_MLP
* 결과 : 0.3957622486
* keras로 작성하는게 더 용이해서 keras로 Sequential model구축하고 baseline과 동일한 layer 유지

```python
import numpy as np
import pandas as pd
import tensorflow as tf

import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
unique_labels = train_df['type'].unique()
label_dict = {val: i for i, val in enumerate(unique_labels)}
i2lb = {v:k for k, v in label_dict.items()}
```

```python
scaler = StandardScaler()
labels = train_df['type']
train_df = train_df.drop(columns=['fiberID', 'type']) # fiberID는 1000개의 categorical feature이며, 이 커널에서는 무시합니다.
```

```python
_mat = scaler.fit_transform(train_df)
train_df = pd.DataFrame(_mat, columns=train_df.columns, index=train_df.index)
```

```python
train_x = train_df
train_y = labels.replace(label_dict)
```

```python
import keras 
from keras import models
from keras import layers
from keras import optimizers
```

```python
optimizer = optimizers.Adam(lr = 0.0003)
```

```python
model = models.Sequential()
model.add(layers.Dense(256*4, activation='elu',input_dim = (train_x.shape[1])))
model.add(layers.Dropout(0.1))
model.add(layers.Dense(256*4, activation='elu'))
model.add(layers.Dense(256*3, activation='elu'))
model.add(layers.Dropout(0.15))
model.add(layers.Dense(256*3, activation='elu'))
model.add(layers.Dense(256*2, activation='elu'))
model.add(layers.Dropout(0.1))
model.add(layers.Dense(256*2, activation='elu'))
model.add(layers.Dropout(0.1))
model.add(layers.Dense(256*2, activation='elu'))
model.add(layers.Dropout(0.15))
model.add(layers.Dense(256*1, activation='elu'))
model.add(layers.Dense(256*1, activation='elu'))
model.add(layers.Dropout(0.15))
model.add(layers.Dense(128, activation='elu'))
model.add(layers.Dropout(0.1))
model.add(layers.Dense(64, activation='elu'))
model.add(layers.Dense(32, activation='elu'))
model.add(layers.Dense(19, activation='softmax'))
```

```python
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_x.values,
          train_y.values,
          batch_size=256*3,
          validation_split=0.1,
          epochs=200)
```

```python
test_ids = test_df.index.values
test_df = test_df.drop(columns=['fiberID'])
test_df = pd.DataFrame(scaler.transform(test_df), columns=test_df.columns, index=test_df.index)
```

```python
pred_mat = model.predict(test_df)
```

```python
submission = pd.DataFrame(pred_mat, index=test_df.index)
submission = submission.rename(columns=i2lb)
# submission = pd.concat([test_ids, submission], axis=1)
submission = submission[sample_submission_df.columns]
submission.to_csv("submission.csv")
```

# Baseline_RF 
* 결과 : 0.5469542991

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
train = pd.read_csv(path +'train.csv', index_col=0)
test = pd.read_csv(path +'test.csv', index_col=0)
sample_submission = pd.read_csv(path +'sample_submission.csv', index_col=0)
```

```python
column_number = {}
for i, column in enumerate(sample_submission.columns):
    column_number[column] = i
    
def to_number(x, dic):
    return dic[x]

train['type_num'] = train['type'].apply(lambda x: to_number(x, column_number))
```

```python
train_x = train.drop(columns=['type', 'type_num'], axis=1)
train_y = train['type_num']
test_x = test
```

```python
forest = RandomForestClassifier(n_estimators=100, n_jobs=-1)
forest.fit(train_x, train_y)
```

```python
pi_df = pd.DataFrame(forest.feature_importances_, index= train_x.columns,columns= ['importance']).sort_values('importance',ascending = False)
pi_df.head()
```

```python
y_pred = forest.predict_proba(test_x)
```

```python
submission = pd.DataFrame(data=y_pred, columns=sample_submission.columns, index=sample_submission.index)
submission.to_csv('submission.csv', index=True)
```
