1. 배경

최근 인류에게 다가온 빅데이터라는 단어는 우주와 천문학에게 낯설지 않습니다.  
찰나의 순간에도 우주는 천문학적인 양의 데이터를 생산해왔고, 오래 전부터 천문학자들은 우주를 관측했으며, 그 방대함에 비례하는 데이터를 수집 및 분석했기 때문입니다. 
슬론 디지털 천체 관측(Sloan Digital Sky Survey: 이하 SDSS)는 세계적 천체 관측 프로젝트로, 우주에 대한 천문학적인 규모의 데이터를 수집하고 있습니다.   
이곳에서 수집한 데이터는 약 6,000개 논문에 사용되었고, 25만 회 이상 인용되었을 정도로 천문학에 큰 기여를 했습니다.  
점점 거대해지는 규모에 따라 데이터 처리에는 머신러닝과 딥러닝 기법이 활용되기 시작했습니다.  
여전히 우주에는 다양한 미지의 이야기가 남아있고, 오늘날 인간은 하늘에서 많은 데이터를 얻어낼 정도로 발전했습니다.  
이 데이터를 분석하여 어쩌면 드러나지 않은 규칙이 여러분의 손끝에서 밝혀질 수 있습니다.   
새로운 알고리즘을 통해 우주의 비밀을 찾아주세요!

        최종순위는 참가자가 선택한 파일로 채점되므로, 참가자는 자신이 채점 받고 싶은 제출 파일을 최종적으로 선택해야 합니다.

3. 컬럼 설명
    * type(Source type) : 천체의 분류 예측을 해야하는 변수
    * psfMag(Point spread function magnitudes): 먼 천체를 한 점으로 가정하여 측정한 빛의 밝기입니다.
    * fiberMag(Fiber magnitudes) : 3인치 지름의 광섬유를 사용하여 광스펙트럼을 측정합니다. 광섬유를 통과하는 빛의 밝기입니다.
    * petroMag(Petrosian Magnitudes) : 은하처럼 뚜렷한 표면이 없는 천체에서는 빛의 밝기를 측정하기 어렵습니다. 천체의 위치와 거리에 상관없이 빛의 밝기를 비교하기 위한 수치입니다.
    * modelMag(Model magnitudes) : 천체 중심으로부터 특정 거리의 밝기입니다.
    * fiberID : 관측에 사용된 광섬유의 구분자

```python
def draw_boxplot(selected_col, data, y_value, nrow=2, ncol=3, figsize=(12,12)):
    plt.figure(figsize=figsize)
#     plt.suptitle(title, fontsize=15)
    for i in range(0,len(selected_col)):
        col_name = selected_col[i]
        sample_df = data[(data[col_name] >0)&(data[col_name] <60)]
        plt.subplot(nrow,ncol,i+1)
        g = sns.boxplot(x=y_value, y=selected_col[i], data= sample_df)
        g.set_title(selected_col[i], fontsize=10)
        g.set_ylabel("")
        plt.xticks(rotation=90)
        plt.axhline(sample_df.psfMag_u.median(),c= 'r')
        print(f"{col_name} 제외된 Data수 : {data.shape[0]-sample_df.shape[0]}")
    plt.tight_layout()
    return 
```

```python
def draw_boxplot_t(selected_col, data, nrow=2, ncol=3, figsize=(12,12)):
    plt.figure(figsize=figsize)
#     plt.suptitle(title, fontsize=15)
    for i in range(0,len(selected_col)):
        col_name = selected_col[i]
        sample_df = data[(data[col_name] >0)&(data[col_name] <60)]
        plt.subplot(nrow,ncol,i+1)
        g = sns.boxplot(y = selected_col[i], data= sample_df)
        g.set_title(selected_col[i], fontsize=10)
        g.set_ylabel("")
        plt.xticks(rotation=90)
        plt.axhline(sample_df.psfMag_u.median(),c= 'r')
        print(f"{col_name} 제외된 Data수 : {data.shape[0]-sample_df.shape[0]}")
    plt.tight_layout()
    return 
```

### type별 변수확인

```python
def type_check(data, nrow=7, ncol=3, figsize=(25,40),min_value = 0, max_value = 60):
    plt.figure(figsize=figsize)
    type_name = train_df.type.unique()
    for i,name in enumerate(type_name): 
        sample = data[data.type == name]
        ori_count = sample.shape[0]
        sample.drop(['type','fiberID'],axis= 1, inplace= True)
        
        #melt
        sample_melt = pd.melt(sample)
        sample_melt = sample_melt[(sample_melt.value>min_value)&(sample_melt.value<max_value)]
        sample_min = sample_melt["variable"].value_counts().min()
        print(f"타입이름 : {name}, 기존 : {ori_count}, 제외후 : {sample_min}, 차이 : {ori_count-sample_min}")
        #draw graph
        plt.subplot(nrow,ncol,i+1)
        g = sns.boxplot(x= "variable", y= "value",data = sample_melt)
        g.set_title(name, fontsize=10)
        g.set_ylabel("")
        plt.xticks(rotation= 90)
        plt.ylim((min_value,max_value))
        plt.axhline(20,c= 'r')
        plt.axhline(sample_melt.value.median(),c= 'b')
    plt.tight_layout()
```

### 타입별 추가변수 확인

```python
def add_minus_feature(data,test = False):
    '''
    근접변수에 대해서만 컬럼생성 ex) 'fiberMag_u-g', 'fiberMag_g-r', 'fiberMag_r-i', 'fiberMag_i-z'
    각 mag에서 가장 높은값의 filter를 받아오는 "~~mag_MAX"생성
    결과 38개로 증가 
    '''
    from itertools import combinations
    n = 0
    m = 0
    for count in range(5,21,5):
        s = 2
        if test == True :
            s = 1
        selected = data.columns[s:].values[n:count]
        mag = str.split(selected[0],'_')[0]
#         data[mag+"_MAX"] = data[selected].apply(lambda x : str.split(np.argmax(x),'_')[1],axis = 1)
        for i in range(0,len(selected)):        
            if i>0: 
                name_1st = str.split(m,'_')[1]
                name_2nd = str.split(selected[i],'_')[1]
                data[mag+"_"+name_1st+"-"+name_2nd] = data[m]-data[selected[i]]
            m = selected[i]    
        n=count
    return data
```

# 각변수별 그래프
* 분포 그래프 특정 튀는값때문에 log씌워서 진행

```python
def draw_distplot(selected_col, data, nrow=5, ncol=5, figsize=(15,15)):
    plt.figure(figsize=figsize)
#     plt.suptitle(title, fontsize=15)
    for i in range(0,len(selected_col)):
        col_name = selected_col[i]
        plt.subplot(nrow,ncol,i+1)
        g = sns.distplot(np.log(data[col_name]))
        g.set_title(col_name, fontsize=10)
        g.set_ylabel("")
    plt.tight_layout()
    return 
```

# 추가변수: 각각의 차

```python
def add_minus_feature(data,test = False):
    from itertools import combinations
    n = 0
    for count in range(5,21,5):
        s = 2
        if test == True :
            s = 1
        selected = data.columns[s:].values[n:count]
        mag = str.split(selected[0],'_')[0]
        for combi in list(combinations(selected,2)):
            name_1st = str.split(combi[0],'_')[1]
            name_2nd = str.split(combi[1],'_')[1]
            data[mag+"_"+name_1st+"-"+name_2nd] = data[combi[0]]-data[combi[1]]
        n=count
    return data
```

```python
train_df = add_minus_feature(train_df)
test_df = add_minus_feature(test_df,True)
```
